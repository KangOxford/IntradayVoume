```python
import os
import pandas as pd

# Initialize the directory path and the file patterns
directory_path = '/homes/80/kang/cmem/'
file_pattern_1 = '_d726.csv'
file_pattern_2 = 'summary'

# Create an empty list to store DataFrames
all_dataframes = []

# Loop through each file in the directory
for filename in os.listdir(directory_path):
    if file_pattern_1 in filename and file_pattern_2 in filename:
        # Construct the complete filepath
        filepath = os.path.join(directory_path, filename)
        print(filepath)
        
        # Read the CSV file into a DataFrame
        df = pd.read_csv(filepath, header=None)
        
        # Append the DataFrame to the list
        all_dataframes.append(df)

# Now all_dataframes contains all the DataFrames read from the CSV files
```

```python
long_df = pd.concat(all_dataframes, ignore_index=True,axis=0)
long_df
```
