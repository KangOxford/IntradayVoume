```python
# Sample code to plot turnover and R2 score on dual y-axes using Matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Create a figure and a set of subplots (dual y-axes)
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot 'turnover' and 'x' on the primary y-axis
ax1.plot(d.index, d['turnover'], label='Actual Turnover', color='b')
ax1.plot(d.index, d['x'].astype(float), label='Predicted Turnover', color='g')
ax1.set_xlabel('Date')
ax1.set_ylabel('Turnover', color='b')
ax1.tick_params(axis='y', labelcolor='b')
ax1.legend(loc='upper left')

# Create a second y-axis to plot the R2 score
ax2 = ax1.twinx()
# Since you want only one R2 score per day, we will use the mean R2 score for each unique date
mean_r2_per_day = d.groupby(d.index)['r2_score'].mean()
ax2.scatter(mean_r2_per_day.index, mean_r2_per_day.values, color='r', label='R2 Score')
ax2.set_ylabel('R2 Score', color='r')
ax2.tick_params(axis='y', labelcolor='r')
ax2.legend(loc='upper right')

# Add title and grid
plt.title('Comparison of Actual and Predicted Turnover with R2 Score')
plt.grid(True)

# Show the plot
plt.show()
```
